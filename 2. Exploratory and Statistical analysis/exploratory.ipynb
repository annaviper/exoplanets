```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing clean data

kepler = pd.read_csv('datasets/kepler_1.csv')
kepler.head()
```

```python
kepler_confirmed = kepler[kepler['disposition']=='CONFIRMED']
```

```python
kepler.columns
```

```python
kepler_confirmed = kepler[kepler['koi_disposition']=='CONFIRMED']
```

# Exploratory analysis

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(kepler_confirmed['koi_period'], kepler_confirmed['koi_prad'])
ax.set_xlabel('Orbital period/days')
ax.set_ylabel('Planet radius/xxx')
ax.set_xlim(0,100)
ax.set_ylim(0,40)
plt.show()

# https://www.nasa.gov/ames/kepler/kepler-planet-candidates-july-2015
```
