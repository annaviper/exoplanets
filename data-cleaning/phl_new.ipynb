# Columns

```python
# Dropping columns with pattern in the name
to_drop = [col for col in df.columns if ('_ERROR_MIN' in col) | ('_ERROR_MAX' in col) | ('_MIN' in col) | ('_MAX' in col)]
df.drop(to_drop, axis=1, inplace=True)
```

```python
df.head()
```

```python
# Drop columns we don't want
to_drop = ['P_NAME', 'P_STATUS', 'P_YEAR', 'P_UPDATED', 'S_NAME', 'S_AGE', 'S_ALT_NAMES', 'S_CONSTELLATION', 'S_CONSTELLATION_ABR', 'S_CONSTELLATION_ENG', 'P_DETECTION', 'S_TYPE', 'S_TYPE_TEMP']
df.drop(to_drop, axis=1, inplace=True)
```

```python
# Cleaning column names
df.columns = [x.lower().strip() for x in df.columns]

# Checking if there's columns we don't want
df.columns
```

```python
print(f'Original shape: {original_shape}')
print(f'New shape:      {df.shape}')
```

# Nans

```python
df.drop(3764, inplace=True)
```

```python
# Dropping columns where all values are Nan
df.dropna(axis=1, how='all', inplace=True)
```

```python
# Dropping columns with Nans over X %
df = df.loc[:, df.isnull().sum() < 0.7 * df.shape[0]]
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.isnull# (), cbar=False, cmap='virid# is');
```

```python
# Percentage of NaN values per column
nan_values = pd.DataFrame(((df.isnull().sum()) / df.shape[0]*100).sort_values(ascending=False))
nan_values[nan_values[0] > 0]
```
